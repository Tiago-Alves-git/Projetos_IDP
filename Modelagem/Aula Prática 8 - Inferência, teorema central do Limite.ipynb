# Inferência Estatística

## Problemas motivadores

1) De 50.000 válvulas fabricadas por uma companhia retira-se uma amostra de 400 válvulas e obtém-se a vida média de 800 horas e o desvio padrão de 100 horas. **(a)** qual o intervalo de confiança de 99% para a vida média da população? **(b)** Com que confiança poderia se dizer que a vida média é 800 $\pm $ 0,98? **(c)** Que tamanho deve ter a amostra para que seja de 95% a confiança na estimativa 800 $\pm $ 7,84?

2) A medida dos diâmetros de uma amostra aleatória de 200 rolamentos esféricos, durante uma semana, apresentam média 0,824 polegadas e o desvio padrão 0,042 polegada. Determine os limites de confiança de **(a)** 95% **(b)** 99%

3) Já sabemos que o problema do lançamento de uma moeda pode ser modelado pela Distribuição Binomial. Supondo o lançamento de uma moeda 50 vezes, suponha que tenha ocorrido 36 caras. Esse resultado permite dizer que a moeda é honesta? O que você faria para verificar esta afirmação?

4) Para fins de estudos, duas turmas de uma determinada disciplina foram comparadas. A primeira turma com 40 alunos, nota média 7,4 e desvio padrão 0,8. A segunda turma com 50 alunos, nota média 7,8 e desvio padrão 0,7. Pergunta-se: há diferença significativa entre os aproveitamentos das duas classes, no nível de significância: **(a)** 0,05 (95%) **(b)** 0,01 (99%).

- Problema 1

```python
\documentclass{article}
\usepackage{amsmath}

\begin{document}

\section*{Questão 1: Intervalo de confiança de 99\% para a vida média da população}

Dado que:
\begin{itemize}
    \item Tamanho da amostra (\(n\)) = 400
    \item Vida média da amostra (\(\bar{x}\)) = 800 horas
    \item Desvio padrão (\(\sigma\)) = 100 horas
    \item Nível de confiança = 99\%
\end{itemize}

O intervalo de confiança (IC) para a média populacional é dado pela fórmula:

\[
IC = \bar{x} \pm z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
\]

Para um nível de confiança de 99\%, o valor crítico \(z_{\alpha/2}\) é aproximadamente 2.576. Substituindo os valores:

\[
IC = 800 \pm 2.576 \times \frac{100}{\sqrt{400}} 
\]
\[
IC = 800 \pm 2.576 \times \frac{100}{20} 
\]
\[
IC = 800 \pm 2.576 \times 5
\]
\[
IC = 800 \pm 12.88
\]

Assim, o intervalo de confiança de 99\% é:

\[
(787.12, 812.88)
\]

\section*{Questão 2: Confiança para \(800 \pm 0.98\)}

Queremos encontrar o nível de confiança para afirmar que a média está dentro de \(800 \pm 0.98\).

Usamos a fórmula para o erro padrão da média:

\[
\text{Erro padrão} = \frac{\sigma}{\sqrt{n}} = \frac{100}{\sqrt{400}} = \frac{100}{20} = 5
\]

Agora, calculamos o valor de \(z\) para a margem de erro de 0.98:

\[
z = \frac{\text{margem de erro}}{\text{erro padrão}} = \frac{0.98}{5} = 0.196
\]

Consultando uma tabela da distribuição normal padrão, vemos que \(z = 0.196\) corresponde a uma probabilidade acumulada de 0.0778, ou seja, cerca de 15.56\%.

Portanto, o nível de confiança é aproximadamente \textbf{15.56\%.}

\section*{Questão 3: Tamanho da amostra para uma confiança de 95\% e margem de erro de 7.84}

A fórmula para calcular o tamanho da amostra \(n\) é:

\[
n = \left( \frac{z_{\alpha/2} \times \sigma}{\text{margem de erro}} \right)^2
\]

Para um nível de confiança de 95\%, \(z_{\alpha/2} = 1.96\). Substituindo os valores:

\[
n = \left( \frac{1.96 \times 100}{7.84} \right)^2
\]
\[
n = \left( \frac{196}{7.84} \right)^2
\]
\[
n = (25)^2
\]
\[
n = 625
\]

Portanto, o tamanho da amostra necessária para uma confiança de 95\% e margem de erro de 7.84 é \textbf{625} válvulas.

\end{document}
```

- Problema 2

```python
\documentclass{article}
\usepackage{amsmath}

\begin{document}

\section*{Problema 2: Intervalo de Confiança para diâmetros de rolamentos esféricos}

Dado:
\begin{itemize}
    \item Tamanho da amostra (\(n\)) = 200
    \item Média da amostra (\(\bar{x}\)) = 0,824 polegadas
    \item Desvio padrão (\(\sigma\)) = 0,042 polegadas
\end{itemize}

\section*{(a) Intervalo de confiança de 95\%}

Para calcular o intervalo de confiança de 95\%, usamos a fórmula:

\[
IC = \bar{x} \pm z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
\]

Para um nível de confiança de 95\%, o valor crítico \(z_{\alpha/2} = 1.96\). Substituímos os valores:

\[
IC = 0.824 \pm 1.96 \times \frac{0.042}{\sqrt{200}} 
\]
\[
IC = 0.824 \pm 1.96 \times \frac{0.042}{14.14}
\]
\[
IC = 0.824 \pm 1.96 \times 0.00297
\]
\[
IC = 0.824 \pm 0.00582
\]

Portanto, o intervalo de confiança de 95\% é:

\textbf{\[
(0.8182, 0.8298)
\]}

\section*{(b) Intervalo de confiança de 99\%}

Para o intervalo de confiança de 99\%, o valor crítico \(z_{\alpha/2} = 2.576\). Substituímos os valores na fórmula:

\[
IC = 0.824 \pm 2.576 \times \frac{0.042}{\sqrt{200}}
\]
\[
IC = 0.824 \pm 2.576 \times \frac{0.042}{14.14}
\]
\[
IC = 0.824 \pm 2.576 \times 0.00297
\]
\[
IC = 0.824 \pm 0.00766
\]

Portanto, o intervalo de confiança de 99\% é:

\textbf{\[
(0.8163, 0.8317)
\]}

\end{document}
```
