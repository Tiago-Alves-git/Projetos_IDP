```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson, binom


# Problema 1

# Parâmetros do problema
n = 2000                # número total de indivíduos
p = 0.001               # probabilidade de reação nociva por indivíduo
lambda_value = n * p    # parâmetro da distribuição de Poisson obtido através da multiplicação do numero total x reação por indivíduo

# (a) Probabilidade de exatamente 3 indivíduos sofrerem reação nociva
prob_3 = poisson.pmf(3, lambda_value)

# (b) Probabilidade de mais do que 2 sofrerem tal reação
prob_more_than_2 = 1 - poisson.cdf(2, lambda_value)

print(f"Problema 1 \n")
print(f"(a) Probabilidade de exatamente 3 indivíduos sofrerem reação nociva: {prob_3:.6f} \n")
print(f"(b) Probabilidade de mais do que 2 sofrerem reação nociva: {prob_more_than_2:.6f} \n")
print(f"Tendo em vista que a probabilidade de que mais do que 2 sofram uma reação negativa é de : {prob_more_than_2:.6f} podemos dizer que sim, o rémedio apresentado é seguro. \n")

# Problema 2

# Parâmetros do problema
lambda_value = 3  # média de acidentes por 100 mil veículos

# Valores para os quais calcular as probabilidades
k_values = np.arange(0, 11)

# Calculando as probabilidades para 0 a 10 acidentes
probs = poisson.pmf(k_values, lambda_value)

# Exibir titulo do problema 

print(f"Problema 02 \n")

# Exibindo os resultados
for k, prob in zip(k_values, probs):
    print(f"Probabilidade de haver exatamente {k} acidentes: {prob:.6f} \n")

# Plotando o histograma
plt.bar(k_values, probs, color='skyblue', edgecolor='black')
plt.xlabel('Número de acidentes')
plt.ylabel('Probabilidade')
plt.title('Distribuição de Poisson de Acidentes')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

print(f"Calculando agora a probabilidade de ocorrer entre 4 a 10 acidentes")

# Vou usar o mesmo lambda 3 do problema anterior

# Arranjando para calcular entre 4 e 10
accidents_range = np.arange(4,11)  

# Calculando as probabilidades de haver entre 4 e 10 acidentes
probabilities = poisson.pmf(accidents_range, lambda_value)

# Somando as probabilidades de 4 a 10 acidentes
total_probability = np.sum(probabilities)

print(f"A probabilidade total de ocorrer entre 4 e 10 acidentes é: {total_probability:.4f} \n")

# Plotando o histograma das probabilidades
plt.bar(accidents_range, probabilities, color='skyblue', edgecolor='black')
plt.xlabel('Número de Acidentes')
plt.ylabel('Probabilidade')
plt.title('Distribuição de Probabilidade de Acidentes (4 a 10 Acidentes)')
plt.xticks(accidents_range)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

print(f"""    Na minha visão de mundo, existir a chance de 0.20 para 2 acidentes é algo que considero alto, 0.16 para 4 acidentes também considero alto,
por uma visão pessoal, acredito que seria prudente uma intervenção para procurar uma forma de diminuir principalmente as chances de
2 a 4 acidentes.\n""")

# Problema 3

# Parâmetros do problema
n = 100                 # número total de peças
p = 0.03                # probabilidade de defeito
lambda_value = n * p    # parâmetro da distribuição de Poisson

# Valores para os quais calcular as probabilidades
k_values = np.arange(0, 5)

# Calculando as probabilidades para 0 a 4 peças defeituosas
probs = poisson.pmf(k_values, lambda_value)

# Plotando o histograma
plt.bar(k_values, probs, color='skyblue', edgecolor='black')
plt.xlabel('Número de peças defeituosas')
plt.ylabel('Probabilidade')
plt.title('Distribuição de Poisson para Peças Defeituosas')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

# Exibindo os resultados
for k, prob in zip(k_values, probs):
    print(f"Probabilidade de exatamente {k} peças defeituosas: {prob:.6f} \n")

# Respondendo a Pergunta sobre o problema

print("Tendo em vista que a probabilidade de peças serem defeituosas ser abaixo de 0.25 em todos os casos, acredito\n"
"que a empresa possa se sentir satisfeita. \n")

# Problema 4
print(f"Problema 4\n")
# Parâmetros do problema
n = 10      # número total de peças
p = 0.1     # probabilidade de defeito
lambda_value = n * p

# (a) Usando a Distribuição Binomial
prob_binomial = binom.pmf(2, n, p)

# (b) Usando a Distribuição de Poisson
prob_poisson = poisson.pmf(2, lambda_value)

print(f"(a) Probabilidade de exatamente 2 defeituosas (Binomial): {prob_binomial:.6f}\n")
print(f"(b) Probabilidade de exatamente 2 defeituosas (Poisson): {prob_poisson:.6f}\n")
print(f"\n")
```

## Resumo das Funções `poisson.pmf` e `poisson.cdf` do `scipy.stats`

### Distribuição de Poisson
A distribuição de Poisson é uma distribuição de probabilidade discreta que modela o número de eventos que ocorrem em um intervalo de tempo fixo ou área, assumindo que os eventos ocorrem independentemente e a uma taxa constante.

### 1. `poisson.pmf`

#### O que é?
A função `poisson.pmf` calcula a "Função de Massa de Probabilidade" (PMF) da distribuição de Poisson.

#### Fórmula
\[
$ P(X = k) = \frac{\mu^k e^{-\mu}}{k!} $
\]
- \( k \): Número específico de eventos desejado.
- \( \mu  \): Taxa média de eventos por intervalo.
- \( e \): Constante de Euler (aproximadamente 2.71828).

#### Uso
Retorna a probabilidade de que ocorram exatamente \( k \) eventos quando a taxa média é \( \mu \).

---

### 2. `poisson.cdf`

#### O que é?
A função `poisson.cdf` calcula a "Função de Distribuição Cumulativa" (CDF) da distribuição de Poisson.

#### Fórmula
\[
$ P(X \leq k) = \sum_{i=0}^{k} \frac{\mu^i e^{-\mu}}{i!} $
\]
- \( k \): Número máximo de eventos.
- \( \mu \): Taxa média de eventos por intervalo.

#### Uso
Retorna a probabilidade acumulada de que o número de eventos seja até \( k \), dado que a taxa média é \( \mu \).

---

### Resumo Comparativo
- **`poisson.pmf(k, \mu)`**: Retorna a probabilidade de exatamente \( k \) eventos ocorrerem.
- **`poisson.cdf(k, \mu)`**: Retorna a probabilidade acumulada de que ocorram até \( k \) eventos.

1) Gere números aleatórios, distribuídos de acordo com a distribuição de Poisson. Gere histogramas e mude o valor $\lambda$. Plote nos resultados um ajuste de curvas normal e verifique o acontece com o histograma dos valores distribuídos e acurva teórica.

2) Gere um conjunto de dados aleatórios distribuídos de acordo com a distribuição normal, binomial e de Poisson que lhe permitam concluir quando as 3 distribuições se aproximam.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm, binom

def plot_poisson(lambd):
    # Gerar números aleatórios com distribuição de Poisson
    data = poisson.rvs(mu=lambd, size=10000)

    # Criar histograma
    plt.hist(data, bins=30, density=True, alpha=0.6, label='Poisson')

    # Ajustar curva normal
    x = np.arange(min(data), max(data) + 1)
    plt.plot(x, norm.pdf(x, lambd, np.sqrt(lambd)), 'r', label='Normal')

    plt.legend()
    plt.title(f'Distribuição de Poisson com λ = {lambd}')
    plt.show()

# Exemplos:
plot_poisson(2)
plot_poisson(10)
plot_poisson(30)

def compare_distributions(n, p, lambd):
    # Gerar dados
    data_binom = binom.rvs(n, p, size=10000)
    data_poisson = poisson.rvs(mu=lambd, size=10000)
    data_norm = norm.rvs(loc=n*p, scale=np.sqrt(n*p*(1-p)), size=10000)

    # Plotar histogramas
    plt.hist(data_binom, bins=30, density=True, alpha=0.6, label='Binomial')
    plt.hist(data_poisson, bins=30, density=True, alpha=0.6, label='Poisson')
    plt.hist(data_norm, bins=30, density=True, alpha=0.6, label='Normal')

    plt.legend()
    plt.title('Comparação de Distribuições')
    plt.show()

# Exemplo:
compare_distributions(100, 0.5, 50)
```
