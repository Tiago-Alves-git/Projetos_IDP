# Questão 1

João e Maria estão jogando RPG, um jogo de imaginação em que as regras do mundo são ditadas por jogadas de dados.

Em uma campanha, Maria encontra três itens que permitem aumentar os pontos de vida do jogador. Cada item vai aumentar o número obtido pela soma da jogada de diferentes dados, podendo ser dados de 4, 6, 8, 10 ou 12 faces. As regras para o cálculo do número de pontos de vida para cada item são:

 - Soma da jogada de 1 dado de 12 faces mais 1 dado de 10 faces. Se o resultado for menor que 7, adicionar 1 dado de 4 faces;
 - Soma da jogada de 2 dados de 6 faces e 2 dados de 4 faces. Se o resultado for menor que 7, adicionar 1 dado de 4 faces;
 - Soma da jogada de 6 dados de 4 faces. Se o resultado for menor que 7, adicionar 1 dado de 12 faces

João não sabe qual item escolher, então Maria decide calcular quais seriam os valores esperados e dispersão desses valores para cada um dos itens.

### Instruções

Escreva um programa que simule a jogada de 10 mil dados e calcule a média e desvio-padrão para cadas uma das opções.

Construa um gráfico de barras para cada opção com o número de vezes que cada face foi lançada e ajude João escolher a melhor opção.

```python
##Questão 1

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

## simulações
num_simulacoes = 10000

# Simular jogadas para 1 dado de 12 faces
jogadas_d12 = np.random.randint(1,13, num_simulacoes)

# Simular jogadas para 1 dado de 10 faces
jogadas_d10 = np.random.randint(1,11,num_simulacoes)

# Calcular o valor esperado para cada opção
valor_esperado_d12 = np.mean(jogadas_d12)
valor_esperado_d10 = np.mean(jogadas_d10)

# Somando os valores de d12 e d10
valor_conjunto_d12_d10 = valor_esperado_d10 + valor_esperado_d12

if (valor_conjunto_d12_d10 < 7) :
    # Simular jogadas para 1 dado de 04 faces
    jogadas_d4 = np.random.randint(1,5,num_simulacoes)
    valor_esperado_d4 = np.mean(jogadas_d4)
    valor_conjunto_d12_d10_d4 = valor_conjunto_d12_d10 + valor_esperado_d4
    # Graficos para jogadas com os 3 dados
    plt.subplot(3,3,3)
    plt.hist([jogadas_d10, jogadas_d12, jogadas_d4], bins=np.arange(1,14) - 0.5, edgecolor='black', rwidth=0.8)
    plt.xticks(np.arange(1,13))
    plt.title(f'Valor conjunto esperado dos 3 dados, 10,12,04 faces = {valor_conjunto_d12_d10_d4}')
    plt.xlabel('Resultado')
    plt.ylabel('Frequencia')
    plt.tight_layout()
    plt.show()
else :
        plt.subplot(2,2,2)
        plt.hist([jogadas_d10, jogadas_d12],bins=np.arange(1,14) ,edgecolor='black', rwidth=0.8)
        plt.xticks(np.arange(1,13))
        plt.title(f'Valor conjunto esperado dos 2 dados, 10 e 12 faces = {valor_conjunto_d12_d10}')
        plt.xlabel('Resultado')
        plt.ylabel('Frequencia')
        plt.tight_layout()
        plt.show()

# Simular jogadas para 1 dado de 06 faces
jogadas_d06 = np.random.randint(1,7, num_simulacoes)
jogadas_d06_2 = np.random.randint(1,7,num_simulacoes)
soma_jogadas_d6 = jogadas_d06 + jogadas_d06_2

# Simular jogadas para 1 dado de 04 faces
jogadas_d04 = np.random.randint(1,5,num_simulacoes)
jogadas_d04_2 = np.random.randint(1,5,num_simulacoes)
soma_jogadas_d4 = jogadas_d04 + jogadas_d04_2

# Calcular o valor esperado para cada opção
valor_esperado_d06 = np.mean(soma_jogadas_d6)
valor_esperado_d04 = np.mean(soma_jogadas_d4)

# Somando os valores de d04 e d06
valor_conjunto_d04_d06 = valor_esperado_d06 + valor_esperado_d04

if (valor_conjunto_d04_d06 < 7) :
    # Simular jogadas para 1 dado de 04 faces
    jogadas_d4 = np.random.randint(1,5,num_simulacoes)
    valor_esperado_d4 = np.mean(jogadas_d4)
    valor_conjunto_d04_d06_d4 = valor_conjunto_d04_d06 + valor_esperado_d4
    # Graficos para jogadas com os 3 dados
    plt.subplot(3,3,3)
    plt.hist([soma_jogadas_d6, soma_jogadas_d4, jogadas_d4], bins=np.arange(1,14) - 0.5, edgecolor='black', rwidth=0.8)
    plt.xticks(np.arange(1,13))
    plt.title(f'Valor conjunto esperado dos 3 dados, de 06,04,04 faces = {valor_conjunto_d04_d06_d4}')
    plt.xlabel('Resultado')
    plt.ylabel('Frequencia')
    plt.tight_layout()
    plt.show()
else :
        plt.subplot(2,2,2)
        plt.hist([soma_jogadas_d6, soma_jogadas_d4],bins=np.arange(1,14) ,edgecolor='black', rwidth=0.8)
        plt.xticks(np.arange(1,13))
        plt.title(f'Valor conjunto esperado dos 2 dados de 06 e 04 faces = {valor_conjunto_d04_d06}')
        plt.xlabel('Resultado')
        plt.ylabel('Frequencia')
        plt.tight_layout()
        plt.show()


#  - Soma da jogada de 6 dados de 4 faces. Se o resultado for menor que 7, adicionar 1 dado de 12 faces

soma_jogadas_d4_6x = 0

for _ in range(6) :
   jogadas_d04 = np.random.randint(1,5, num_simulacoes)
   soma_jogadas_d4_6x += jogadas_d04

# Calcular o valor esperado da soma de 4 faces, 06 vezes
valor_esperado_d04_6x = np.mean(soma_jogadas_d4_6x)

if (valor_esperado_d04_6x < 7) :
    # Simular jogadas para 1 dado de 12 faces
    jogadas_d12 = np.random.randint(1,13, num_simulacoes)
    valor_esperado_d12 = np.mean(jogadas_d12)
    valor_conjunto_d04_6x_d12 = valor_esperado_d04_6x + valor_esperado_d12
    # Graficos para jogadas com os 2 dados
    plt.subplot(2,2,2)
    plt.hist([soma_jogadas_d4_6x, jogadas_d12], bins=np.arange(1,14) - 0.5, edgecolor='black', rwidth=0.8)
    plt.xticks(np.arange(1,13))
    plt.title(f'Valor conjunto esperado dos 2 dados, de 04x6 + 12x1 faces = {valor_conjunto_d04_6x_d12}')
    plt.xlabel('Resultado')
    plt.ylabel('Frequencia')
    plt.tight_layout()
    plt.show()
else :
        plt.subplot(1,1,1)
        plt.hist([soma_jogadas_d4_6x],bins=np.arange(1,14) ,edgecolor='black', rwidth=0.8)
        plt.xticks(np.arange(1,13))
        plt.title(f'Valor conjunto esperado dos 6 dados de 04 faces = {valor_esperado_d04_6x}')
        plt.xlabel('Resultado')
        plt.ylabel('Frequencia')
        plt.tight_layout()
        plt.show()



```

De acordo com os gráficos, apesar de certa mudança para cada vez que o codigo é rodado, todas mostram que rodar 06 dados de 04 faces é a melhor escolha para o caso de João.
